```python
%py
# Variables
origin_volume_path = '/Volumes/bike_store/resource_schema/resouce/origin/'
external_bronze_path = 'abfss://container20260122@datalake20260121.dfs.core.windows.net/01-bronze/'

# dbutils.fs.ls(origin_volume_path)
# dbutils.fs.ls(external_bronze_path)
origin_folder = dbutils.fs.ls(origin_volume_path)
```

```python
# Writing databricks origin volume CSVs into azure bronze layer as Delta files
for file in origin_folder:
    df = spark.read.csv(file.path, header=True, inferSchema=True)

    df.write.format('delta').mode('overwrite').option('mergeSchema', True).save(external_bronze_path + file.name)
```
